```python
#analysing the distribution of data
platform_counts = df['Platform'].value_counts()

plt.figure(figsize=(5, 5))
sns.countplot(data=df, x='Platform', order=platform_counts.index)
plt.xlabel('Platform')
plt.ylabel('Number of Clicks')  # Adjust the label to reflect what is being counted
plt.title("Platform")
plt.xticks(rotation=90)  # Rotate x-axis labels if necessary for better readability
plt.show()
```

```python
#Will plot a piediagram
graph=df['Audience_Segment'].value_counts()
plt.figure(figsize=(10,5))
plt.pie(graph, labels= graph.index, autopct='%1.1f%%')
plt.title('Audience_Segment')

plt.show()
```
